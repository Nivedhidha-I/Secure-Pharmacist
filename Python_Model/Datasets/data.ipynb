```python
import pandas as pd

# Loading Dataset
def load_data(d):
    data = pd.read_csv(d, low_memory=False, encoding='utf-8')
    lowercase = lambda x: str(x).lower()
    data.rename(lowercase, axis='columns', inplace=True)
    return data
```

```python
Data = load_data("workspaces/Secure-Pharmacist/Python_Model/Datasets/DrugsMungedData.csv")
Data = Data.sort_values("drug_id")
Data
```
